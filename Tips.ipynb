## Notes, warnings, and other admonitions

Jupyter Book allows you to make the learning more interactive and versatile. You can highlight a particular block of text that exists slightly apart from the narrative of your page:

```{note}
Here is a note!
```


is generated by the following:

````md
```{note}
Here is a note!
```
````

Similarly, you may use
```{warning}
Here is a warning!
```
```{admonition} Here's your admonition
Here's the admonition content
```
or use special style blocks:
`````{admonition} This admonition was styled...
:class: tip
With a tip class!
`````
and even dropdown blocks:
```{admonition} Click here!
:class: dropdown
The note body will be hidden!
```

Double clicking any of these admonitions will reveal their format.
